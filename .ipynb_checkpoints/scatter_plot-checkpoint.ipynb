## Scatter Plots in Matplotlib
Scatter plots are used to show the relationship between two variables. Let's assume we are plotting the study hours vs. exam scores for a group of students.

```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]
```

```python
plt.scatter(study_hours,exam_scores)
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Study hour Vs Exam score")
plt.grid(True)
```

## Adding Color and Size

```python
size=[score * 2 for score in exam_scores ]
color=["green" if score> 60 else "red" for score in exam_scores]
plt.scatter(study_hours,exam_scores,s=size,c=color)
plt.grid(True)
plt.title('Colored & Sized Scatter Plot')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.grid(True)
```

 ## Using a Colormap

```python
import numpy as np
array_score=np.array(exam_scores)
plt.scatter(study_hours,exam_scores,c=array_score,cmap="plasma")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Study vs Marks plot")
plt.annotate("s1",(study_hours[0],exam_scores[0]),xytext=(5,5),textcoords="offset points")
```

## Adding annotations

```python
plt.scatter(study_hours,exam_scores)
for i in range(len(exam_scores)):
    plt.annotate(f"s{i+1}",(study_hours[i],exam_scores[i]),xytext=(5,5),textcoords="offset points")
    
```

## plotting multiple scatter plot in one graph

```python
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]
 
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]
```

```python
plt.scatter(class_a_hours,class_a_scores,label="Class A",color="red")
plt.scatter(class_b_hours,class_b_scores,label="Class B",color="blue")
plt.legend()
plt.grid(True)


for i in range(len(class_a_scores)):
    plt.annotate(f"S{i+1}_A",(class_a_hours[i],class_a_scores[i]),xytext=(5,5),textcoords="offset points")
for i in range(len(class_b_scores)):
    plt.annotate(f"S{i+1}_B",(class_b_hours[i],class_b_scores[i]),xytext=(5,5),textcoords="offset points")
plt.margins(0.1)

plt.title('Scatter Plot: Class A vs Class B')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
```
