## Subplots in Matplotlib
Subplots allow you to show multiple plots in a single figure, side-by-side or in a grid layout. This is helpful when comparing different datasets or aspects of the same data.



```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y1=[i*2 for i in x]
y2=[i**2 for i in x]
y3=[i**0.5 for i in x]
y4=[10-i for i in x]
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("Double")
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("Square")
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("Sqaure_root")
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("Minus by 10")
plt.tight_layout()
```

## Using plt.subplots() for Clean Code

```python
fig,axs=plt.subplots(1,2,figsize=(10,4))
axs[0].plot(x,y1)
axs[0].set_title("Double")
axs[1].plot(x,y2)
axs[1].set_title("Square")
fig.suptitle("Comparison of operation graphs",fontsize=14)
fig.tight_layout()
fig.subplots_adjust(top=0.8)  # So title doesn't overlap
fig.savefig('my_plots.png')
```
