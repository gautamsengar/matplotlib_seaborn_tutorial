# Matplotlib Introduction

Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations. It is widely used in data analysis, machine learning, and scientific computing for plotting graphs such as line plots, bar charts, scatter plots, histograms, and more. The most commonly used module in Matplotlib is pyplot, which provides a simple interface similar to MATLAB for creating plots easily and efficiently.

```python
import matplotlib.pyplot as plt
```

### Real Data Example: Sachin Tendulkar’s Runs Over Time

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.plot(years,runs)
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Sachin Tendulkar's yearly runs")
```

### Multiple Lines in One Plot

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0] 
plt.plot(years,kohli,label="Kohli")
plt.plot(years,sehwag,label="Sehwag")
plt.legend()
```

### Using Format Strings , Color and Line Style Arguments

```python
plt.plot(years,kohli ,'bo--',label="Kohli") 
plt.plot(years,sehwag ,'g^:' ,label="Sehwag")
plt.plot(years,runs,color='orange',linestyle='-',label = "Tendulkar")
plt.legend()
```

### Line Width and Layout Tweaks

```python
plt.plot(years,runs,linewidth=10)
plt.grid(True)
plt.style.use("ggplot")
```
